Дополнительные базовые классификаторы:

Показатель качества модели – доля верных ответов (accuracy) 

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import StackingClassifier

# Логистическая регрессия с полиномиальными признаками
log_reg = LogisticRegression(max_iter=10000)
log_reg.fit(X_train, y_train)

# Метод опорных векторов
svm_clf = SVC()
svm_clf.fit(X_train, y_train)

# StackingClassifier
stack_clf = StackingClassifier(estimators=[
    ('lr', log_reg),
    ('svc', svm_clf)
], final_estimator=DecisionTreeClassifier(max_depth=4))
stack_clf.fit(X_train, y_train)
```

Комбинированный классификатор: StackingClassifier

```python
# Оценка точности комбинированного классификатора
stack_accuracy = stack_clf.score(X_test, y_test)
print(f"Точность StackingClassifier: {stack_accuracy}")
```
